回归分析


Dep.Variable: y 因变量
Model：OLS 最小二乘模型
Method: Least Squares 最小二乘
No. Observations: 样本数据的数量
X_factor Residuals：残差自由度(degree of freedom of residuals)
X_factor Model：模型自由度(degree of freedom of model)
Covariance Type：nonrobust 协方差阵的稳健性
R-squared：R 判定系数
Adj. R-squared: 修正的判定系数
F-statistic： 统计检验 F 统计量
Prob (F-statistic): F检验的 P值
Log likelihood: 对数似然
 
coef：自变量和常数项的系数，b1,b2,...bm,b0
std err：系数估计的标准误差
t：统计检验 t 统计量
P>|t|：t 检验的 P值
[0.025, 0.975]：估计参数的 95%置信区间的下限和上限
Omnibus：基于峰度和偏度进行数据正态性的检验
Prob(Omnibus)：基于峰度和偏度进行数据正态性的检验概率
Durbin-Watson：检验残差中是否存在自相关
Skewness：偏度，反映数据分布的非对称程度
Kurtosis：峰度，反映数据分布陡峭或平滑程度
Jarque-Bera(JB)：基于峰度和偏度对数据正态性的检验
Prob(JB)：Jarque-Bera(JB)检验的 P值。
Cond. No.：检验变量之间是否存在精确相关关系或高度相关关系

聚类分析

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from matplotlib import pyplot as plt
```

```python
X_cluster = np.array(data_set.iloc[:,1:15])
Y_cluster = np.array(data_set.iloc[:,0:1])
Y_cluster_list = list(Y_cluster)
```

离差平方和层次聚类

```python
z = linkage(X_cluster,'ward')
```

```python
fig = plt.figure(figsize=(25,10))
plt.rcParams['font.sans-serif'] = ['SimHei']
Y_labels = []
for i in range(0,len(X)):
    Y_labels.append(Y_cluster_list[i])

dn = dendrogram(z, labels = Y_labels, orientation='right')
plt.show()
```

```python
cluster_w = fcluster(z, t=6, criterion='maxclust') 
cluster = cluster_w.reshape(31,1)
ward_cluster = np.hstack((Y_cluster,cluster))
ward_cluster
```

最短距离法层次聚类

```python
z = linkage(X_cluster,'single')
```

```python
fig = plt.figure(figsize=(25,10))
plt.rcParams['font.sans-serif'] = ['SimHei']
Y_labels = []
for i in range(0,len(X)):
    Y_labels.append(Y_cluster[i])

dn = dendrogram(z, labels = Y_labels, orientation='right')
plt.show()
```

```python
cluster_min = fcluster(z, t=1, criterion='inconsistent')  #cluster = fcluster(z, t=5, criterion='maxclust')
cluster = cluster_min.reshape(31,1)
min_cluster = np.hstack((Y_cluster,cluster))
min_cluster
```

最长距离法层次聚类

```python
z = linkage(X_cluster,'complete')
```

```python
fig = plt.figure(figsize=(25,10))
plt.rcParams['font.sans-serif'] = ['SimHei']
Y_labels = []
for i in range(0,len(X)):
    Y_labels.append(Y_cluster[i])

dn = dendrogram(z, labels = Y_labels, orientation='right')
plt.show()
```

```python
cluster_max = fcluster(z, t=1, criterion='inconsistent') #cluster = fcluster(z, t=5, criterion='maxclust')
cluster = cluster_max.reshape(31,1)
max_cluster = np.hstack((Y_cluster,cluster))
max_cluster
```

贝叶斯判别分析

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB_x = X_cluster
NB_y = cluster_w
indices = np.random.permutation(len(NB_x))  # 产生一个随机打乱的0-31的一维数组

NB_train_province = np.arange(21).reshape([21,1])
NB_train_province = NB_train_province.astype(np.str_)
NB_test_province = np.arange(10).reshape([10,1])
NB_test_province = NB_test_province.astype(np.str_)
for i in range(0,21):
    NB_train_province[i] = Y_cluster_list[indices[i]]
for i in range(21,31):
    NB_test_province[i-21] = Y_cluster_list[indices[i]]

NB_x_train = NB_x[indices[:-10]]  # 随机选取21个样本作为训练数据集
NB_y_train = NB_y[indices[:-10]]  # 选取这21个样本的标签作为训练数据集的标签
NB_x_test = NB_x[indices[-10:]]  # 剩下的10个样本作为测试数据集
NB_y_test = NB_y[indices[-10:]]  # 把剩下10个样本对应标签作为测试数据集的标签

NB_test_true = np.arange(20).reshape([10,2])
NB_test_true = NB_test_true.astype(np.str_)
for i in range(21,31):
    a = []
    a.append(ward_cluster[indices[i]][0])
    a.append(ward_cluster[indices[i]][1])
    NB_test_true[i-21][0] = ward_cluster[indices[i]][0]
    NB_test_true[i-21][1] = ward_cluster[indices[i]][1]

NB_train = np.arange(42).reshape([21,2])
NB_train = NB_train.astype(np.str_)
for i in range(0,21):
    a = []
    a.append(ward_cluster[indices[i]][0])
    a.append(ward_cluster[indices[i]][1])
    NB_train[i][0] = ward_cluster[indices[i]][0]
    NB_train[i][1] = ward_cluster[indices[i]][1]
```

```python
bayes = MultinomialNB()  
bayes.fit(NB_x_train, NB_y_train)

NB_y_predict = bayes.predict(NB_x_test)
NB_y_predict = NB_y_predict.reshape(10,1)

NB_test_predict = np.hstack((NB_test_province, NB_y_predict))

# 输出训练数据集
print("训练数据集为：\n", NB_train)
# 输出测试数据集的预测标签
print('测试数据集的预测标签为：\n ', NB_test_predict)
# 输出测试数据集的正确标签，以方便对比
print("测试数据集的正确标签为：\n", NB_test_true)
# 调用该对象的打分方法，计算出准确率
score = bayes.score(NB_x_test,NB_y_test,sample_weight=None)
print("准确率为：",score)  # 输出准确率计算结果
```
